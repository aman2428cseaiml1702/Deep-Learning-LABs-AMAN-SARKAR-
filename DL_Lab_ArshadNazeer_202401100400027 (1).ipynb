Experiment 2: Creating and Displaying Tensors

```python
a = tf.constant(5)
print(a)

v = tf.constant([10, 20, 30])
print(v)

m = tf.constant([
    [1, 2],
    [3, 4]
])

print(m)
```

Experiment 3: Arithmetic Operations on Tensor

```python
a = tf.constant([10, 20])
b = tf.constant([5, 15])

s = tf.add(a, b)
print(s)

d = tf.subtract(a, b)
print(d)

p = tf.multiply(a, b)
print(p)

d = tf.divide(a, b)
print(d)
```

Experiment 4: Displaying info of a Tensor

```python
tensor = tf.constant(50)

print("Data type:", tensor.dtype)

print("Shape:", tensor.shape)

print("Number of dimensions:", tensor.ndim)
```

Experiment 5: converting a numpy array to tensor

```python
import numpy as np

np_array = np.array([[1, 2, 3],
                     [4, 5, 6]])

tensor = tf.convert_to_tensor(np_array)

print("Tensor:")
print(tensor)
```
